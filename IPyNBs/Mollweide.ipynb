```python
%matplotlib inline
import numpy as np, matplotlib.pyplot as plt
from astropy import coordinates as coord, units as u, constants, cosmology
from mpl_toolkits.basemap import Basemap
from mpl_toolkits.mplot3d import axes3d
from IPython.html.widgets import interact, interactive, fixed
from IPython.html import widgets
```

```python
ra, dec, z = np.loadtxt('SDSSgalaxy_local_SG_DR9.txt', usecols = (0, 1, 2), unpack=True) #300000 galaxies catalogue
ra_isol, dec_isol, z_isol = np.loadtxt('isolated_galaxies.txt', usecols = (0, 1, 2), unpack=True)
ra_pair, dec_pair, z_pair = np.loadtxt('isolated_pairs.txt', usecols = (0, 1, 2), unpack=True)
ra_trip, dec_trip, z_trip = np.loadtxt('isolated_triplets.txt', usecols = (0, 1, 2), unpack=True)
zstart, zrange, dec0, alpha0 = .030, .005, 30., .2
cosmo = cosmology.FlatLambdaCDM(H0=70., Om0=0.3)
d = constants.c.to('km/s') * z / cosmo.H0
```

```python
def plotSkymap(Isolated=True, Pairs=False, Triplets=False, z0=zstart, rang0=zrange, Transparency=alpha0):
    
    plt.figure(figsize=(15, 7))
    plt.subplot(111)
    m = Basemap(projection='moll', lon_0=180, resolution='c', celestial=True)
    m.drawmeridians(np.arange(0.,420.,60.)) ; m.drawparallels(np.arange(-90.,120.,30.), labels=[1,0,0,1])
    m.drawmapboundary(fill_color='w')
    
    x, y = m(ra, dec)
    x_isol, y_isol = m(ra_isol, dec_isol)
    x_pair, y_pair = m(ra_pair, dec_pair)
    x_trip, y_trip = m(ra_trip, dec_trip)
    
    cond = (z >= z0) & (z < z0+rang0)
    cond_isol = (z_isol >= z0) & (z_isol < z0+rang0)
    cond_pair = (z_pair >= z0) & (z_pair < z0+rang0)
    cond_trip = (z_trip >= z0) & (z_trip < z0+rang0)
    b, = plt.plot(x[cond], y[cond], 'ko', ms=1, alpha=Transparency)
    c, = plt.plot(x_isol[cond_isol],y_isol[cond_isol], 'ro', ms=4, alpha=.7, visible=Isolated)
    d, = plt.plot(x_pair[cond_pair],y_pair[cond_pair], 'go', ms=4, alpha=.7, visible=Pairs)
    e, = plt.plot(x_trip[cond_trip],y_trip[cond_trip], 'bo', ms=4, alpha=.7, visible=Triplets)
    plt.show()
    
    fig = plt.figure(figsize=(6, 6))
    ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.8], polar=True)
    zeta = [0, 1, 1, 0]
    cond_dec = (dec > -1.) & (dec <= 1.)
    cond_z = (z[cond_dec] >= z0) & (z[cond_dec] < z0 + rang0)
    ax2.scatter(np.radians(ra[cond_dec]-95.), z[cond_dec], c=z[cond_dec], s=1, marker=',', lw=0, alpha=.1)
    ax2.scatter(np.radians(ra[cond_dec][cond_z]-95.), z[cond_dec][cond_z], c='r', s=1, marker=',', lw=0, alpha=1)
    ax2.bar(0., rang0, width=2*np.pi, bottom=z0, color='r', edgecolor='r', linewidth=0, alpha=.4)
    ax2.set_rmax(.1)
    #ax2.set_xticks(fontsize=6)
    plt.setp(ax2.get_xticklabels(), fontsize=10)
    plt.setp(ax2.get_yticklabels(), fontsize=8)    
    plt.show()
```

```python
interact(
    plotSkymap,
    Isolated=True, 
    Pairs=False, 
    Triplets=False, 
    z0=widgets.FloatSliderWidget(min=0.0, max=0.1, step=0.01, value=zstart),
    rang0=widgets.FloatSliderWidget(min=0.0, max=0.1, step=0.01, value=zrange),
    Transparency=widgets.FloatSliderWidget(min=0.0, max=1.0, step=0.1, value=alpha0)
)
```
